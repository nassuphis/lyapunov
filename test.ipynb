```python
import os
def add_suffix_number(path: str, n: int) -> str:
    base, ext = os.path.splitext(path)
    return f"{base}_{n}{ext}"
```

```python
add_suffix_number("test.jpg",1)
```

```python
import sys
from pathlib import Path
parent = Path("test.ipynb").resolve().parent.parent
sys.path.insert(0, str(parent))
import re as regex
from rasterizer import colors

def get_tri_palettes(palette_regex, maxp):
    print(f"tri palette search:{palette_regex}")
    pat = regex.compile(palette_regex)
    out = []
    for k in colors.COLOR_TRI_STRINGS.keys():
        if pat.search(k):
            print(f"found palette: {k}")
            out.append(k)
            if len(out) >= maxp:
                break
    return out

get_tri_palettes("^(?!(rg$|gc$|bp$)).+",10)
```
