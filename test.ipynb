```python
import sys
from pathlib import Path
parent = Path("test.ipynb").resolve().parent.parent
sys.path.insert(0, str(parent))
from rasterizer import colors
import random
import itertools
import re as regex
import importlib
importlib.reload(colors)
print(colors.__file__)

def random_tri_pair_set(name:str, n: int, seed: int):
    keys = list(colors.COLOR_TRI_STRINGS.keys())
    pat = regex.compile(name)
    allowed = [k for k in keys if pat.fullmatch(k)]

    m = len(allowed)
    if m < 2 or n <= 0:
        return []

    # total unique unordered pairs
    total = m * (m - 1) // 2
    if n > total:
        n = total  # or raise ValueError

    rng = random.Random(seed)

    # sample unique unordered pairs
    pairs = rng.sample(list(itertools.combinations(allowed, 2)), n)

    # optional: randomize orientation in the output string
    out = []
    for a, b in pairs:
        if rng.random() < 0.5:
            a, b = b, a
        out.append(f"{a}:{b}")

    return out

print(colors.COLOR_TRI_STRINGS["beauty_gucci_taupe_blush"])
random_tri_pair_set("^leaves.*",50,10)
```

```python
import subprocess

def read_spec_exiftool(path: str) -> str:
    out = subprocess.check_output(
        ["exiftool", "-s3", "-XMP-lyapunov:spec", path],
        text=True,
    )
    if out.strip():
        return out.strip()

    # fallback to EXIF
    out = subprocess.check_output(
        ["exiftool", "-s3", "-UserComment", path],
        text=True,
    )
    return out.strip()

read_spec_exiftool("nn14loc91/nn14loc91_00249_autolvl.jpg")
```

```python
import functions

functions.NS.keys()

import inspect

def nargs(fn):
    sig = inspect.signature(fn)
    return sum(
        p.kind in (p.POSITIONAL_ONLY, p.POSITIONAL_OR_KEYWORD)
        and p.default is inspect._empty
        for p in sig.parameters.values()
    )

def describe_functions(NS: dict[str, callable]) -> str:
    parts = []
    for name, fn in NS.items():
        try:
            n = nargs(fn)
            parts.append(f"{name}() takes {n} arg{'s' if n != 1 else ''}")
        except (TypeError, ValueError):
            parts.append(f"{name}() has unknown arity")
    return ", ".join(parts)

describe_functions(functions.NS)
```
